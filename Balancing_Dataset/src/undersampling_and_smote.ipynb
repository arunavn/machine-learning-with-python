```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('cuisines.csv')
data.shape
```

```python
data.head()
```

```python
x = data['cuisine'].value_counts()
import seaborn as sns
sns.barplot(x)
```

```python
data = data.drop('Unnamed: 0', axis = 1)
```

```python
data.head()
```

```python
features_df = data.drop('cuisine', axis= 1)
labels_df = data.cuisine
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
tranformed_feature , transformed_labels = oversample.fit_resample(features_df, labels_df)
```

```python
new_data = pd.concat([tranformed_feature, transformed_labels], axis=1)
```

```python
new_data.head()
```
